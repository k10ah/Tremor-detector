```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score
import joblib
```

```python
# =============================
# Paths
# =============================
feature_file = r"D:\hypoglycemia\features.csv"
model_file = r"D:\hypoglycemia\rf_model.pkl"
```

```python
k_folds = 5
```

```python
# =============================
# Load features
# =============================
data = pd.read_csv(feature_file)
X = data.drop(columns=['label'])
y = data['label']
```

```python
# =============================
# Model and K-Fold Cross Validation
# =============================
clf = RandomForestClassifier(n_estimators=300, random_state=42)
```

```python
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)
scores = cross_val_score(clf, X, y, cv=kf, scoring='accuracy')
```

```python
print(f"{k_folds}-Fold Cross Validation Accuracy Scores: {scores}")
print(f"Mean Accuracy: {scores.mean():.4f} ± {scores.std():.4f}")
```

```python
# =============================
# Train on full dataset & save model
# =============================
clf.fit(X, y)
joblib.dump(clf, model_file)
print(f"Model saved to: {model_file}")
```
